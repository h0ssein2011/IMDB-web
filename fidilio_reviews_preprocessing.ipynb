```python
import requests
from bs4 import BeautifulSoup
url = "https://fidilio.com/restaurants/darya_baz/%D8%AF%D8%B1%DB%8C%D8%A7-%D8%A8%D8%A7%D8%B2/"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
comment_body = soup.find(class_='tab review-paginate')
comment=comment_body.find_all(class_='text-limitation show all mobile-hd')
#number of comment 
reviewlist=soup.find('div',id='review-list')
re_count=reviewlist.find(class_='reviews-count')
nazar_number=re_count.get_text()
nazar_number=nazar_number.replace(' ','')
nazar_number=nazar_number.replace('تعدادنظرات:','')
s=int(nazar_number)
comments=[]
for i in range(s):
    comments.append(comment[i].get_text())
comments
```

```python
#delete of numbers
import re
import nltk.data
comments=str(comments)
import re
remove_digits =True
pattern = r'[^الف-ی آ-ی0-9\s]' if not remove_digits else r'[^الف-یآ-ی\s]'
comments = re.sub(pattern, '', comments)
comments
```

```python
#حذف علامت ها و اسپیس ها
import string
exclude = set(string.punctuation)
comments = ''.join(ch for ch in comments if ch not in exclude)
comments
```

```python
h=comments.splitlines()
for i in h:
    if i=='':
        h.remove('')
comments=str(h)
comments= comments.replace('  ','')
comments
```

```python
# نرمالسازی با کتابخانه هضم
from __future__ import unicode_literals
from hazm import *
normalizer = Normalizer()
comments=normalizer.normalize(comments)
comment_words=word_tokenize(comments)
```

```python
len(comment_words)
```

```python
#حذف استپ ورد
from hazm import stopwords_list
for i in comment_words:
    if i in stopwords_list():
        comment_words.remove(i)
```
